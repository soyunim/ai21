```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
```

```python
x,y = make_circles(n_samples=500,factor=0.5,noise=0.05)
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
y_k_pred = kmeans.fit_predict(x)
```

```python
dbscan = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
y_db_pred = dbscan.fit_predict(x)
```

```python
plt.title("K means")
plt.scatter(x[y_k_pred==0,0], x[y_k_pred==0,1], label="cluster1")
plt.scatter(x[y_k_pred==1,0], x[y_k_pred==1,1], label="cluster2")
plt.legend(loc=(1.1,0.0))
plt.show()
```

```python
plt.title("DBSCAN")
plt.scatter(x[y_db_pred==0,0], x[y_db_pred==0,1], label="cluster1")
plt.scatter(x[y_db_pred==1,0], x[y_db_pred==1,1], label="cluster2")
plt.legend(loc=(1.1,0.0))
plt.show()
```
